```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Linkedin_data.csv')
df
```

```python
df['d1'] = df['Dates'].str.split(',')
df['d2'] = df['Covid_Zones'].str.split(',')
df=df.explode(['d1','d2']).reset_index()
df=df.drop(['Dates'], axis=1)
df=df.drop(['Covid_Zones'],axis=1)
df
```

```python
dict={'d1':'Dates',
      'd2':'Covid_Zones'}
df.rename(columns = dict, inplace = True)
df
```

```python
test=df
```

```python
#test['Dates']=test['Dates'].astype(str)
test['Dates']=pd.to_datetime(df['Dates'])
y = test.pivot(index=['Area'], columns = ['Dates'],values=['Covid_Zones'])
y
```

```python
y.to_excel('Final_Covid_Report.xlsx',sheet_name='Covid Report')
```
